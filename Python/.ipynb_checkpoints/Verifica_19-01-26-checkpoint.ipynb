Esercizio 2: Analisi Statistica e Scelta Soglie 

Per definire cosa sia un "Bacino Idrico" in questa specifica immagine, dobbiamo campionare i valori dei pixel nelle zone d'acqua evidenti.

    Crea una maschera temporanea  (sicuramente acqua, es. B > 200).

    Utilizzando questa selezione, calcola e stampa i valori medi degli altri fenomeni fisici in quelle zone:

    Altitudine Media: Calcola la media del canale R nei pixel selezionati.

    Pendenza Media: Calcola la media del canale G nei pixel selezionati.

    Suggerimento: Usa la funzione np.mean() sui pixel filtrati.

    Decisione Operativa:

    Soglia Altitudine (Min/Max): (es. Media +/- una tolleranza(50))

    Soglia Pendenza (Max): (es. Media + una tolleranza(50))

```python
R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]

avg_R = np.mean(R)
avg_G = np.mean(G)
avg_B = np.mean(B)
print(f"Media R: {avg_R}, Media G: {avg_G}, Media B: {avg_B}")

mask_acqua = B>180
avg_acqua_R = np.mean(R[mask_acqua])
avg_acqua_G = np.mean(G[mask_acqua])
```

Utilizza le soglie decise al punto precedente per creare le maschere definitive.

    mask_alt: Isola i pixel con l'altitudine corretta (basata sulla media calcolata).

    mask_pend: Isola i pixel con pendenza compatibile (basata sulla media calcolata).

    mask_acqua: Isola i pixel con acqua presente (es. > 180 o > 200).

    Combina le maschere usando l'operatore logico & per trovare i bacini idrici.

```python
mask_alt = (R>avg_acqua_R-51) & (R<avg_acqua_R+51)
mask_pend = (G>avg_acqua_G-51) & (G<avg_acqua_G+51)
mask_finale = mask_acqua & mask_alt & mask_pend
```
