```python
import numpy as np
import matplotlib.pyplot as plt
```

importa librerie

```python
scacchiera = np.zeros((800, 800))
```

crea scaciera

```python
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            scacchiera[i * 100:(i + 1) * 100, j * 100:(j + 1) * 100] = 1
```

ciclo sigma

```python
plt.imshow(scacchiera, cmap='gray')
```

```python
plt.imsave('scacchiera_800x800.tiff', scacchiera, cmap='gray')
```
