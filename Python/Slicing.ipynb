Creazione array

```python
import numpy as np

# Creazione dlla prima colonna (0, 10, 20, 30, 40, 50) e reshape in Colonna 6x1
# arange(start, stop, step)
moltiplicatori_riga = np.arange(0, 60, 10).reshape(6, 1) # reshape(new_shape) cambia forma 

# Creazione della prima riga (0, 1, 2, 3, 4, 5)
offset_colonna = np.arange(6)
# Esegue somma elemento per elemento (vettorializzazione) 
matrice_6x6 = moltiplicatori_riga + offset_colonna

print(matrice_6x6)
```

Selezione Arancione 3, 4

```python
selezione_arancione = matrice_6x6[0, 3:5]
print(selezione_arancione)
```

Selezione Rossa 2 12 22 32 42 52

```python
selezione_rossa = matrice_6x6[:, 2]
print(selezione_rossa)
```

Selezione Verde 20 22 24 40 42 44

```python
selezione_verde = matrice_6x6[2:5:2, 0:5:2]
print(selezione_verde)
```

Selezione Grigia 44 45 54 55

```python
selezione_grigia = matrice_6x6[4:6, 4:6]
print(selezione_grigia)
```
