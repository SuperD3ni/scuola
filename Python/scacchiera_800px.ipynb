```python
import numpy as np
import matplotlib.pyplot as plt
scacchiera=np.zeros((800,800))
scacchiera
```

```python
for i in range(0,9,2):
    scacchiera[i*100:i*100+100, (i+1)*100+100:(i+1)*100]=255
    scacchiera[(i+1)*100+100:(i+1)*100, i*100:i*100+100]=255
scacchiera
```

```python
plt.imshow(scacchiera,cmap="gray")
```
